```python
import numpy as np
import pandas as pd

from sklearn.metrics import accuracy_score
```

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
# Print the keys of the dataset
print("Keys of Iris dataset: ", iris_dataset.keys())
```

```python
# Print the values of the dataset
print("Target names: ", iris_dataset['target_names'])
print("Feature names: ", iris_dataset['feature_names'])
print("Data: ", iris_dataset['data'])
print("Target: ", iris_dataset['target'])
```

```python
print("Description of the dataset", iris_dataset['DESCR'])
```

```python
print("Shape of data", (iris_dataset['data'].shape))
print("Shape of target", (iris_dataset['target'].shape))
```

```python
print("Sample of data", iris_dataset.data[:5])
print("")
print("Sample of target", iris_dataset.target)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (iris_dataset.data, iris_dataset.target, random_state = 0)
```

```python
print("X_train shape", X_train.shape)
print("X_test shape", X_test.shape)
print("y_train shape", y_train.shape)
print("y_test shape", y_test.shape)
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot of the X_train data
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

```python
# Create a list of all feature names
feature_names = iris_dataset.feature_names

# Create a 4x4 grid of subplots
fig, axes = plt.subplots(4, 4, figsize=(15, 15))

# Loop over all pairs of features and create a scatter plot for each pair
for i in range(len(feature_names)):
    for j in range(len(feature_names)):
        if i != j:
            ax = axes[i, j]
            ax.scatter(X_train[:, i], X_train[:, j], c=y_train)
            ax.set_xlabel(feature_names[i])
            ax.set_ylabel(feature_names[j])

# Adjust the spacing between subplots and display the plot
plt.subplots_adjust(wspace=0.4, hspace=0.4)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=2)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
print("Test set score", model.score(X_test,y_test))
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns


# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Visualize the confusion matrix using a heatmap
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d", xticklabels=iris_dataset.target_names, yticklabels=iris_dataset.target_names)
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.show()
```
